````python
def rating_i (income):#—Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –Ω–æ–≤–æ–≥–æ –ø–µ—Ä–µ–∏–º–µ–Ω–æ–≤–∞–Ω–Ω–æ–≥–æ —Ä–µ–π—Ç–∏–Ω–≥–∞
    try:
        if income == 'ao':
            return '18+'
        elif income == 'e':
            return 'all'
        elif income == 'e10+':
            return '10+'
        elif income <= 'ec':
            return '3+'
        elif income == 'k-a':
            return 'all'
        elif income == 'm':
            return '17+'
        elif income == 'rp':
            return 'no_rating'
        elif income == 't':
            return '13+'
    except:
        return 'no_rating'
````

- –í–ª–∏—è–µ—Ç –ª–∏ —Ä–µ–π—Ç–∏–Ω–≥ ESRB –Ω–∞ –ø—Ä–æ–¥–∞–∂–∏ –≤ –æ—Ç–¥–µ–ª—å–Ω–æ–º —Ä–µ–≥–∏–æ–Ω–µ?

````python
#–§—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–æ—Ä–∏—Å–æ–≤–∫–∏ –≥—Ä–∞—Ñ–∏–∫–æ–≤ top_5_in_regions
def top_5_in_regions(groupby, all_sales, ax):
    pivot = actual_data.groupby(groupby).agg({all_sales: 'sum'}).sort_values(by=all_sales, ascending=False)[:5]
    title_dict = {'na_sales': 'North America Sales', 'eu_sales': 'Europe Sales', 'jp_sales': 'Japan Sales'}
    color_dict = {'na_sales': 'red', 'eu_sales': 'gray', 'jp_sales': 'forestgreen'}
    plot = pivot.plot(kind='bar', ec='black', title=title_dict[all_sales], ax=axes[ax], 
                   fontsize=18, color=color_dict[all_sales], rot=20)
    plot.legend(prop={'size':17})
    plot.set_xlabel('')
    plot.title.set_size(20)
````
